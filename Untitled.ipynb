### Data processing
Wines with <6.5 are bad wines while others are good wine

```python
wine_df['quality']=wine_df.quality.apply(lambda x:1 if x>=7 else 0)
```

```python
wine_df['quality'].value_counts()
```

```python
X=wine_df.drop('quality',axis=1)
y=wine_df['quality']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

## Model Training

#### Logistic Regression

```python
log_reg=LogisticRegression()
```

```python
log_reg.fit(X_train,y_train)
```

```python
log_reg_y_pred=log_reg.predict(X_test)
```

```python
accuracy_score(log_reg_y_pred,y_test)
```

```python
confusion_matrix(log_reg_y_pred,y_test,labels=log_reg.classes_)
```

```python
print(classification_report(log_reg_y_pred,y_test))
```

### Decision Tree

```python
dectree=DecisionTreeClassifier()
```

```python
dectree.fit(X_train,y_train)
```

```python
dec_y=dectree.predict(X_test)
```

```python
accuracy_score(dec_y,y_test)
```

```python
print(classification_report(dec_y,y_test))
```

```python
style.use('classic')
cm=confusion_matrix(y_test,dec_y,labels=dectree.classes_)
disp=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=dectree.classes_)
disp.plot()
```

```python
confusion_matrix(dec_y,y_test,labels=dectree.classes_)
```

### Random Forest

```python
rf=RandomForestClassifier()
```

```python
rf.fit(X_train,y_train)
```

```python
rf_y=rf.predict(X_test)
```

```python
accuracy_score(rf_y,y_test)
```

```python
print(classification_report(rf_y,y_test))
```

```python
style.use('classic')
cm=confusion_matrix(y_test,rf_y,labels=rf.classes_)
disp=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=rf.classes_)
disp.plot()
```

```python
confusion_matrix(rf_y,y_test,labels=rf.classes_)
```
